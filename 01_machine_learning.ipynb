## Parameter, Hyperparameter, Parameter Tuning, Model Tuning

- To increase model prediction performance

**Parameter:** inner coefficients in the model - regression coefficients

**Hyperparameter:** exterior values that are found by the used by trying different values in order to optimize the model

**Parameter tuning:** Setting parameters for optimization; it can be used for parameters and hyperparameters

**Model tuning:** Finding model parameters by using hyperparameters, feature engineering without hyperparameter, choosing variable, choosing model, improving the model by hyperparameters are the parts of model tuning
